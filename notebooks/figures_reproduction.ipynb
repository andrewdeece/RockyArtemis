## 2. Frequency Dependence Plot (Figure 2) — Regenerable Example

```python
# Example regeneration of Figure 2 (Debye response)
omega = np.logspace(-7, 2, 1000)
tau_ent = 0.01
alpha0 = 1.0

real_part = alpha0 / (1 + (omega * tau_ent)**2)
imag_part = alpha0 * omega * tau_ent / (1 + (omega * tau_ent)**2)

plt.figure(figsize=(10, 6))
plt.semilogx(omega, real_part, 'b-', label='Re[α_eff]')
plt.semilogx(omega, imag_part, 'r--', label='Im[α_eff]')
plt.axvline(1/tau_ent, color='green', linestyle='--', label='Artemis window crossover')
plt.xlabel('Frequency ω (Hz)')
plt.ylabel('Screening Function α_eff')
plt.title('Frequency Dependence of Screened Dilaton Coupling (Reproducible)')
plt.legend()
plt.grid(True)
plt.show()
```
